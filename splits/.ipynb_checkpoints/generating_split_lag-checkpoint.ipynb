```python
import glob
import random
```

```python
root_path = '/workspace/grape_video_compressed_4'
video_dir_list_temp = glob.glob(root_path + '/***/**')
random.shuffle(video_dir_list_temp)
video_dir_list_train = video_dir_list_temp[:int(len(video_dir_list_temp)*0.8)]
video_dir_list_test = video_dir_list_temp[int(len(video_dir_list_temp)*0.8):]
frame_path_list_train_temp = []
frame_path_list_test_temp = []
frame_path_list_train = []
frame_path_list_test = []
```

```python
for direc in video_dir_list_train:
    temp_frame_path_list = glob.glob(direc + '/*jpg')
    #frame_path_list_train_temp.append(temp_frame_path_list)
    frame_path_list_train_temp = frame_path_list_train_temp + temp_frame_path_list

for direc in video_dir_list_test:
    temp_frame_path_list = glob.glob(direc + '/*jpg')
    #frame_path_list_test_temp.append(temp_frame_path_list)
    frame_path_list_test_temp = frame_path_list_test_temp + temp_frame_path_list
```

```python
import os
frame_interval = 3
for index in range(len(frame_path_list_train_temp)):
    frame_name = frame_path_list_train_temp[index].split('/')[-1]
    root_path = frame_path_list_train_temp[index].replace(frame_name , '')
    frame_index = int(frame_name.replace('frame', '').replace('.jpg', ''))
    
    adjacent_previous_frame_path = root_path + '/frame' + str(frame_index - frame_interval) + '.jpg'
    adjacent_following_frame_path = root_path + '/frame' + str(frame_index + frame_interval) + '.jpg'
    
    if(os.path.exists(adjacent_previous_frame_path) and os.path.exists(adjacent_following_frame_path)):
        frame_path_list_train.append( root_path + ' ' + str(frame_index))
        
    else:
        print("False: " + root_path + ' ' + str(frame_index))
```

```python
import os
for index in range(len(frame_path_list_test_temp)):
    frame_name = frame_path_list_test_temp[index].split('/')[-1]
    root_path = frame_path_list_test_temp[index].replace(frame_name , '')
    frame_index = int(frame_name.replace('frame', '').replace('.jpg', ''))
    
    adjacent_previous_frame_path = root_path + '/frame' + str(frame_index - frame_interval) + '.jpg'
    adjacent_following_frame_path = root_path + '/frame' + str(frame_index + frame_interval) + '.jpg'
    
    if(os.path.exists(adjacent_previous_frame_path) and os.path.exists(adjacent_following_frame_path)):
        #frame_path_list[index] = root_path + ' ' + str(frame_index)
        frame_path_list_test.append( root_path + ' ' + str(frame_index))
        #print("True: " + train_path_list[index])
        
    else:
        print("False: " + root_path + ' ' + str(frame_index))
```

```python
random.shuffle(frame_path_list_train)
random.shuffle(frame_path_list_test)

with open('./omni_grape_interval_3/train_files.txt', 'w') as f:
    for d in frame_path_list_train:
        f.write("%s\n" % d)
        
with open('./omni_grape_interval_3/val_files.txt', 'w') as f:
    for d in frame_path_list_test:
        f.write("%s\n" % d)
```
